```python
import pandas as pd
import scanpy as sc
import numpy as np
import collections
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
import os
```

```python
SAMPLE = "YF-2818_8HetClonesHashtagMix_hLLL_multiome"
by_clone_cache_dir = f'/data/peer/landm/Projects/sail-dogma/yi-clone-samples/cache/{SAMPLE}/by_clone/'
fig_dir = f'/data/peer/landm/Projects/sail-dogma/yi-clone-samples/notebooks/figures/2024_03_27/'
```

```python
hto_dict = {
    'A0251': '0%',
    'A0252': '25%',
    'A0253': '40%',
    'A0254': '40%',
    'A0255': '58%',
    'A0256': '60%',
    'A0257': '75%',
    'A0258': '91%'
}

cpal = dict(zip(hto_dict.keys(), sns.color_palette()))
```

### For each clone, we compute and plot the single-cell heteroplasmy

```python
count_dfs = {}
for clone in os.listdir(by_clone_cache_dir):
    out_dir = f'{by_clone_cache_dir}{clone}/mito/single-cell-coverage_split_3000_bp_frags/'

    count_df = pd.read_csv(f"{out_dir}counts_df.csv", index_col = 0)
    before = count_df.shape[0]
    
    # filtering cells for at least 20 average coverage
    count_df = count_df[count_df.mean(axis = 1) >= 20]
    print(f"Removing {clone}: {before - count_df.shape[0]} cells with avg coverage below 20.")
    count_dfs[clone] = count_df
```

```python
mito_len = 16569

start = mito_len - 8556
end = mito_len - 5109

# originally 1396 - 4514 (FISH_PROBE REGION), adjusted to be same length as del region
common_start = 1231
common_end = 4678 

print(f'common region: {common_start} - {common_end}')
print(f'deleted region: {start} - {end}')
```

### Next, we can plot the distribution of heteroplasmy across all cells within the clone

## Figure S6b

```python
fig, axs = plt.subplots(2, 4, figsize=(20,10), sharex =True, 
                        # sharey = True
                       )

ordered_keys = ['A0251', 'A0252', 'A0253', 'A0254', 'A0255', 'A0256', 'A0257', 'A0258',]

for ax, clone in zip(axs.reshape(-1), ordered_keys):
    # count coverage in deleted region
    del_counts = count_dfs[clone][[str(x) for x in range(start - 1, end)]].sum(axis = 1)
    common_counts = count_dfs[clone][[str(x) for x in range(common_start - 1, common_end)]].sum(axis = 1)

    heteroplasmy = del_counts/common_counts
    heteroplasmy.loc[heteroplasmy > 1] = 1

    heteroplasmy = 1 - heteroplasmy
    
    sns.histplot(heteroplasmy, bins = np.linspace(0, 1, 26), color = cpal[clone], ax = ax)
    ax.set(title = f'{clone} ({hto_dict[clone]} heteroplasmy)')
    # shared_barcodes = list(set(count_dfs[clone].index).intersection(set(adata.obs_names)))
    # adata = adata[shared_barcodes, :]
    # adata.obs['pct_from_deleted_region'] = pct_del.loc[shared_barcodes]
    # adata.obs['log_total_mito_cov'] = np.log(total_counts.loc[shared_barcodes])
# plt.yscale('log')
# plt.savefig(f'{fig_dir}8_clone_heteroplasmy.svg', dpi = 300, bbox_inches = 'tight')
plt.show()
```

## Figure S6a

```python
fig, axs = plt.subplots(2, 4, figsize=(20,10), sharex =True, 
                        sharey = True
                       )

ordered_keys = ['A0251', 'A0252', 'A0253', 'A0254', 'A0255', 'A0256', 'A0257', 'A0258',]

for ax, clone in zip(axs.reshape(-1), ordered_keys):
    df = count_dfs[clone]
    cov = pd.DataFrame(df.mean(), columns = ['count_norm'])
    cov['bp'] = cov.index.astype(int)
    
    ax = sns.lineplot(x=cov['bp'], y=cov['count_norm'], 
             color= cpal[clone], label = clone,
              ax=ax)
    ax.set(ylabel = 'Average Coverage', title = f'{clone} ({hto_dict[clone]} heteroplasmy)')
    ax.get_legend().remove()
    
# plt.suptitle(f'Distribution of Fragments with length greater than {threshold}')
# plt.suptitle(f'Distribution of Fragments with length 1000-3000')
plt.savefig(f'{fig_dir}8_clone_coverage.svg', dpi = 300, bbox_inches = 'tight')
plt.show()
```

## We plot coverage of deleted region per cell

```python
fig, axs = plt.subplots(2, 4, figsize=(20,10), sharex =True, 
                        sharey = True
                       )

ordered_keys = ['A0251', 'A0252', 'A0253', 'A0254', 'A0255', 'A0256', 'A0257', 'A0258',]


max_x = -1
for ax, clone in zip(axs.reshape(-1), ordered_keys):
    # count coverage in deleted region
    del_counts = count_dfs[clone][[str(x) for x in range(start - 1, end)]].sum(axis = 1)
    del_counts = del_counts/(end - start)
    if np.max(del_counts.values) > max_x:
        max_x = np.max(del_counts.values)
        
for ax, clone in zip(axs.reshape(-1), ordered_keys):
    del_counts = count_dfs[clone][[str(x) for x in range(start - 1, end)]].sum(axis = 1)
    del_counts = del_counts/(end - start)
    sns.histplot(del_counts, bins = np.linspace(0, max_x, 30), color = cpal[clone], ax = ax)
    ax.set(title = f'{clone} ({hto_dict[clone]} heteroplasmy)', xlabel = 'Coverage')
# plt.yscale('log')
plt.suptitle('Average Coverage of Deleted Region')
plt.show()
```
